```python
#importing packages
import numpy as np
import pandas as pd

```

```python
#Reading csv file
dataset = pd.read_csv("https://raw.githubusercontent.com/vidakpop/simple-insurance-policy-recommendation-sytem/main/policy.csv")

dataset
```

```python
#separating dependent and independent columns
X = dataset.iloc[:, [0, 1]].values
y = dataset.iloc[:, 2].values

```

```python
#Splitting dataset into test data and train data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

```

```python
#Fitting DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(X_train, y_train)

```

```python
#predicting for x_test data
y_pred = model.predict(X_test)

# Model Accuracy, how often is the classifier correct?
from sklearn import metrics
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))

```

```python

# Take user input for age and income
age = float(input("Enter the age: "))
income = float(input("Enter the income: "))

# Create a list with the user's input
user_input = [[age, income]]

# Make predictions using the model
predictions = model.predict(user_input)

# Display the predictions
print("Predicted Policy: ", predictions[0])
```
